```python
# install pyspark
!pip install pyspark
import os
import sys
import re
os.environ["PYSPARK_PYTHON"] = sys.executable
os.environ["PYSPARK_DRIVER_PYTHON"] = sys.executable
from pyspark.sql.functions import col, avg, udf, format_number, LIT
from pyspark.sql.types import DoubleType
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("IT_Job_Market_Analytics") \
    .config("spark.sql.execution.arrow.pyspark.enabled", "true") \
    .getOrCreate()
```

menghitung rata-rata gaji nasional industri IT sesuai rumpun

```python
def ekstrak_gaji(teks_gaji, cari_maksimal=False):
    if not teks_gaji or "rahasia" in teks_gaji.lower():
        return None

    if cari_maksimal:
        match = re.search(r"-\s*Rp\s*([\d,\.]+)", teks_gaji)
    else:
        match = re.search(r"Rp\s*([\d,\.]+)", teks_gaji)
    if not match:
        return None

    angka_mentah = match.group(1)
    angka_bersih = re.sub(r"[\.,]", "", angka_mentah)

    try:
        nilai = float(angka_bersih)
        if "juta" in teks_gaji.lower():
            if "," in angka_mentah or "." in angka_mentah:
                return nilai * 100000
            else:
                return nilai * 1000000
        return nilai
    except:
        return None

udf_gaji_min = udf(lambda x: ekstrak_gaji(x, cari_maksimal=False), DoubleType())
udf_gaji_max = udf(lambda x: ekstrak_gaji(x, cari_maksimal=True), DoubleType())
df_gaji_final = df_spark.withColumn("salary_min", udf_gaji_min(col("salary_range"))) \
                        .withColumn("salary_max", udf_gaji_max(col("salary_range"))) \
                        .withColumn("salary_average", (col("salary_min") + col("salary_max")) / 2)

print("Rata-rata Gaji Berdasarkan Posisi (Top 10)")
df_tabel_gaji = df_gaji_final.groupBy("job_title") \
    .agg(avg("salary_average").alias("Rata_Rata_Gaji")) \
    .filter(col("Rata_Rata_Gaji").isNotNull()) \
    .orderBy(col("Rata_Rata_Gaji").desc()) \
    .select(
        col("job_title"),
        format_number(col("Rata_Rata_Gaji"), 0).alias("Rata_Rata_Gaji_Rp")
    ) \
    .limit(10) \
    .toPandas()
df_tabel_gaji
```

pengelompokan kluster pekerjaan menjadi beberapa kelompok besar

```python
df_rumpun = df_spark.withColumn(
    "job_family",
    when(lower(col("job_title")).contains("data") |
         lower(col("job_title")).contains("analyst") |
         lower(col("job_title")).contains("scientist") |
         lower(col("job_title")).contains("intelligence") |
         lower(col("job_title")).contains("ai") |
         lower(col("job_title")).contains("learning"), "Data & Artificial Intelligence")

    .when(lower(col("job_title")).contains("developer") |
          lower(col("job_title")).contains("engineer") |
          lower(col("job_title")).contains("programmer") |
          lower(col("job_title")).contains("architect") |
          lower(col("job_title")).contains("frontend") |
          lower(col("job_title")).contains("backend"), "Software Engineering")

    .when(lower(col("job_title")).contains("devops") |
          lower(col("job_title")).contains("cloud") |
          lower(col("job_title")).contains("sre") |
          lower(col("job_title")).contains("infrastructure"), "Cloud & DevOps")

    .when(lower(col("job_title")).contains("security") |
          lower(col("job_title")).contains("cyber") |
          lower(col("job_title")).contains("pentester") |
          lower(col("job_title")).contains("tester"), "Cyber & Software Security")

    .when(lower(col("job_title")).contains("support") |
          lower(col("job_title")).contains("helpdesk") |
          lower(col("job_title")).contains("network") |
          lower(col("job_title")).contains("system admin"), "IT Infrastructure & Support")

    .otherwise("Other IT / Management Roles")
)

print("Distribusi Lowongan Pekerjaan\nBerdasarkan Rumpun Pekerjaan IT")

df_analisis_rumpun = df_rumpun.groupBy("job_family") \
    .count() \
    .orderBy(col("count").desc()) \
    .toPandas()
df_analisis_rumpun
```
